```python
# Pandas is used for data manipulation
import pandas as pd

# NumPy is used for numerical computations
import numpy as np
```

```python
# Read the CKD dataset into a DataFrame
df = pd.read_csv("kidney_disease.csv")

# Display the first five records
print(df.head())
```

```python
# Count missing values in each column
print(df.isnull().sum())
```

```python
# Calculate missing percentage
missing_percent = (df.isnull().sum() / len(df)) * 100

print(missing_percent)
```

```python
# Check duplicate records
print("Duplicates:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

print("New Shape:", df.shape)
```

```python
# Display data types
print(df.dtypes)
```

```python
# Select numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Fill missing values with median
for column in numerical_columns:
    df[column].fillna(df[column].median(), inplace=True)
```

```python
# Select categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Fill missing values with mode
for column in categorical_columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
# Check whether missing values remain
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for column in categorical_columns:
    df[column] = encoder.fit_transform(df[column])
```

```python
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

feature_columns = df.drop("classification", axis=1).columns

df[feature_columns] = scaler.fit_transform(df[feature_columns])
```

```python
print(df.head())
```

```python
# Save the processed dataset for future experiments
df.to_csv("processed_ckd.csv", index=False)

print("Processed dataset saved successfully.")
```
