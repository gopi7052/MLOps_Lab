```python
# Import required libraries
import pandas as pd
import joblib

# Import Machine Learning algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
# Load the fitted preprocessing pipeline created in Lab 3

preprocessor = joblib.load("preprocessing_pipeline.pkl")

print("Preprocessing pipeline loaded successfully.")
```

```python
# Load processed training and testing datasets

X_train = pd.read_csv("X_train_processed.csv")

X_test = pd.read_csv("X_test_processed.csv")

# Load target variables

y_train = pd.read_csv("y_train.csv").squeeze()

y_test = pd.read_csv("y_test.csv").squeeze()

print("Training Data Shape :", X_train.shape)
print("Testing Data Shape :", X_test.shape)
```

```python
# Create Logistic Regression model

lr = LogisticRegression(max_iter=1000)

# Train the model

lr.fit(X_train, y_train)

print("Logistic Regression model trained successfully.")
```

```python
# Save Logistic Regression model

joblib.dump(lr, "logistic_regression.pkl")

print("Logistic Regression model saved.")
```

```python
# Create Decision Tree model

dt = DecisionTreeClassifier(random_state=42)

# Train the model

dt.fit(X_train, y_train)

print("Decision Tree model trained successfully.")
```

```python
joblib.dump(dt, "decision_tree.pkl")

print("Decision Tree model saved.")
```

```python
# Create Random Forest model

rf = RandomForestClassifier(random_state=42)

# Train the model

rf.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

```python
joblib.dump(rf, "random_forest.pkl")

print("Random Forest model saved.")
```

```python
# Create KNN model

knn = KNeighborsClassifier()

# Train the model

knn.fit(X_train, y_train)

print("KNN model trained successfully.")
```

```python
joblib.dump(knn, "knn.pkl")

print("KNN model saved.")
```

```python
# Create SVM model

svm = SVC(probability=True)

# Train the model

svm.fit(X_train, y_train)

print("SVM model trained successfully.")
```

```python
joblib.dump(svm, "svm.pkl")

print("SVM model saved.")
```

```python
# Create Naïve Bayes model

nb = GaussianNB()

# Train the model

nb.fit(X_train, y_train)

print("Naive Bayes model trained successfully.")
```

```python
joblib.dump(nb, "naive_bayes.pkl")

print("Naive Bayes model saved.")
```

```python
import os

model_files = [
    "logistic_regression.pkl",
    "decision_tree.pkl",
    "random_forest.pkl",
    "knn.pkl",
    "svm.pkl",
    "naive_bayes.pkl"
]

print("\nSaved Models:")

for model in model_files:
    if os.path.exists(model):
        print(f"✔ {model}")
    else:
        print(f"✘ {model} not found")
```
